```python
gcX = GC(x.seq)
gcY = GC(y.seq)

# print(gcX)
# print(gcY)

# HASIL print gcX & gcY: 
# 37.97277865097148
# 37.97277865097148


mtX = MT.Tm_NN(x.seq)
mtY = MT.Tm_NN(y.seq)

# print(mtX)
# print(mtY)

# HASIL print mtX & mtY:
# 77.76492619134928
# 77.76492619134928

# subplots parameters
# 1 = baris
# 2 = kolum

# code dibawah berarti 1 baris 2 kolum
fig, axs = plt.subplots(1,2)
axs[0].bar(["GC Seq X", "GC Seq Y"], [gcX, gcY]) # ini buat bikin bar chartnya
axs[0].set_title("GC Content") # ini buat set judul dari bar chartnya
axs[1].bar(["MT Seq X", "MT Seq Y"], [mtX, mtY]) # ini buat bikin bar chartnya
axs[1].set_title("Melting Temp (MT)") # ini buat set judul dari bar chartnya

plt.show()
```
