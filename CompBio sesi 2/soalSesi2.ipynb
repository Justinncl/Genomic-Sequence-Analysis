```python
import Bio
from Bio.Seq import Seq

sequence_A = Seq("AGCTTGCAGCGTCCGTTAGCTCGAGTCCAGGACGTTAGTCCTGCAGTC")
sequence_B = Seq("CAGTAAGTTGCCGTTAGCGCGTAGTGCCAGTAAGCGGCTCGTTAGTGG")
```

```python
# Determine the molecular weight of both sequence
from Bio.SeqUtils import molecular_weight as mw

print(f"Molecular Weight of Sequence A: {mw(sequence_A)}")
print(f"Molecular Weight of Sequence B: {mw(sequence_B)}")
```

```python
# Refer to the Wallace rule to calculate the melting point of sequence A and sequence B and plot them using Matplotlib library. (Use Bar Chart)
from Bio.SeqUtils import MeltingTemp as MT 

melt_seqA = MT.Tm_Wallace(sequence_A)
melt_seqB = MT.Tm_Wallace(sequence_B)

print(f"Melting Temp of Sequence A: {melt_seqA}")
print(f"Melting Temp of Sequence B: {melt_seqB}")

import matplotlib.pyplot as plt

# x = yang horizontal -> namanya
# y = yang vertical -> angkanya

plt.bar(height = [melt_seqA, melt_seqB], x = ["Sequence A", "Sequence B"]) 
plt.show()
```

```python
# Calculate the GC content of sequence A and sequence B.
from Bio.SeqUtils import GC

gc_a = GC(sequence_A)
gc_b = GC(sequence_B)

print(f"GC of Sequence A: {gc_a}")
print(f"GC of Sequence B: {gc_b}")

# Calculate the AT content of sequence A and sequence B.
at_a = 100 - gc_a
at_b = 100 - gc_b

print(f"AT of Sequence A: {at_a}")
print(f"AT of Sequence B: {at_b}")

# Plot the GC and AT content of sequence A and sequence B using Matplotlib library (Use Pie Chart).
import matplotlib.pyplot as plt

plt.pie(x = [gc_a, at_a], labels = [round(gc_a, 2), round(at_a, 2)])
plt.legend(["GC", "AT"])
plt.show()

plt.pie(x = [gc_b, at_b], labels = [round(gc_b, 2), round(at_b, 2)])
plt.legend(["GC", "AT"])
plt.show()
```
